```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
df_cat=df.select_dtypes(object)
df_cat.head()
```

```python
df_num=df.select_dtypes(['int64','float64'])
df_num
```

```python
#applying onehotencoding
#gives temporary label
from sklearn.preprocessing import OneHotEncoder
```

```python
#get_dummies is used for onehotencoding
pd.get_dummies(df['Species'])
```

```python
#apply labelencoder for permanent label
from sklearn.preprocessing import LabelEncoder
```

```python
#create object
le=LabelEncoder()
```

```python
#fit_transform function is used for permanent label no.
for col in df_cat:
    df_cat[col]=le.fit_transform(df_cat[[col]])
```

```python
df_cat.head()
```

```python
#no. of each type of species
df['Species'].value_counts()
```

```python
#concat object and numeric
df_new=pd.concat([df_num,df_cat],axis=1)
df_new
```
