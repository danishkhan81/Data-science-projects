```python
#performing Exploratory Data Analysis(EDA) on googplay Dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#load dataset
df=pd.read_csv('gplay.csv')
df.head()
```

```python
#ewmove unwanted column
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.head()
```

```python
#check null
df.isnull().sum()
```

```python
#visualize
sns.heatmap(df.isnull())
plt.show()
```

```python
#check datatypes
df.dtypes
```

```python
#check percentage of null
df.isnull().sum()*100/df.shape[0]
```

```python
#mean 0f rating
m=df['Rating'].mean()
m
```

```python
#fill the null values with mean 0f rating
df['Rating'].fillna(m,inplace=True)
```

```python
df.isnull().sum()
```

```python
#now removing null from Type
df['Type'].value_counts()
```

```python
#we fill null  with free value which are maximum
df['Type'].fillna('Free',inplace=True)
```

```python
#Type column after removing null
df['Type'].value_counts()
```

```python
#while checking datatypes we see the datatype of review and price is object  means there may be some  special chars in init
#we will remove special chars from it
```

```python
#checking unique value in Review column
c=0
for i,data in enumerate(df['Reviews']):
    if not data.isnumeric():
        c=c+1
        print(i," ",data)
print('no. of special chars',c)
```

```python
#check row10472
df.iloc[10472]
```

```python
#here values are nor properly assigned
#we  rerarrange it

df['Reviews'].replace('3.0M','19.0',inplace=True)
```

```python
df.iloc[10472]
```

```python
#checking unique value in price column
c=0
for i,data in enumerate(df['Price']):
    if not data.isnumeric():
        c=c+1
        print(i," ",data)
print('no. of special chars',c)
```

```python
#check unique values in price
df['Price'].unique()
```

```python
#datatype of price is shown object bcoz of $ sign in it.
#we remove $ sign using lstrip
df['Price']=df['Price'].map(lambda x:x.lstrip('$'))
```

```python
#check unique values in price after removing $ sign
df['Price'].unique()
```

```python
df[df['Price']=='Everyone']
```

```python
#here values are nor properly assigned
#we  rerarrange it

#replace null with Everyone value
df['Content Rating'].fillna('Everyone',inplace=True)

#replace price value from Everyone to 0
df['Price'].replace('Everyone','0',inplace=True)

#replace Type value from 0 to Free
df['Type'].replace('0','Free',inplace=True)

#replace installs value from free to 1000+
df['Installs'].replace('Free','1,000+',inplace=True)

#replace size from 1000+ to 3.0M
df['Size'].replace('1,000+','3.0M',inplace=True)
```

```python
#replace rating from 19.0 to 1.9
df['Rating'].replace(19.0,1.9,inplace=True)
```

```python
#category counts
df['Category'].value_counts()
```

```python
df['Category'].replace('1.9','FAMILY',inplace=True)
```

```python
#check row 10472 is properly arranged or not
df.iloc[10472]
```

```python
#checking unique value in size column
df['Size'].unique()
```

```python
#remove M using rstrip to make all values numeric
df['Size']=df['Size'].map(lambda x:x.rstrip('M'))
```

```python
#convert KB to MB
#1024 KB=1 MB
#1MB=1/1024KB
```

```python
#first we remove the k using r strip and then convet KB to MB by dividing k by 1024
df['Size']=df['Size'].map(lambda x:str(round((float(x.rstrip('k'))/1024),1))
                          if x[-1]=='k' else x)

#lamda input :expeession if condition else expression
```
