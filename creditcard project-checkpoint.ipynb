# from above we see data is not balanced.we balance the data using  randon under sampling to balance the data

```python
#select input and output
X=df.drop('Class',axis=1)
Y=df['Class']
```

```python
#train test split the data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
#take all input columns in a list
column=list(X.columns.values)
```

```python
#visualize

non_frauds=df.Class==0
frauds=df.Class==1

plt.figure(figsize=(20,40))
for i,col in enumerate(column):   #enumrate  :used to access index and columnname
    plt.subplot(10,3,i+1)             #subplot(no. of rows, no. of col,index)
    sns.distplot(X[col][non_frauds],color='blue')
    sns.distplot(X[col][frauds],color='red')
    plt.title(col,fontsize=15)
plt.show()
```

```python
#check correlation
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
#call class
from imblearn.under_sampling import RandomUnderSampler

#create object
rus= RandomUnderSampler()
```

```python
#fit_resample() inbuilt method of RandomunderSampler class
X_train_rus,Y_train_rus=rus.fit_resample(X_train,Y_train)
```

```python
#check if not balance
pd.Series(Y_train).value_counts()
```

```python
#check if balance after applying randomundersampling
pd.Series(Y_train_rus).value_counts()
#here we make majority equal to minority
```

```python
#Apply randomundersampling on 30% testing data
X_test_rus,Y_test_rus=rus.fit_resample(X_test,Y_test)
```

```python
#check if not balance
pd.Series(Y_test).value_counts()
```

```python
#check if balance after applying sampling
pd.Series(Y_test_rus).value_counts()
```

```python
#1. Standardization
#always apply scaler on input.
from sklearn.preprocessing import StandardScaler

#fit_transform on training data X_train and 
#transform on testing_data means X_test :
#create object of StandardScaler class
ss=StandardScaler()
X_train_rus=ss.fit_transform(X_train_rus) 
X_test_rus=ss.transform(X_test_rus)
```

```python
#dataset is a classification dataset,
#use baseline model
#create function

def create_model(model):
    model.fit(X_train_rus,Y_train_rus) #train model
    Y_pred=model.predict(X_test_rus) #test model
    print(classification_report(Y_test_rus,Y_pred)) 
    #report always in between actual output and predicted output
    #Confusion matrix
    print(confusion_matrix(Y_test_rus,Y_pred))
    #accuracy score
    print(accuracy_score(Y_test_rus,Y_pred))
    return model
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.metrics import classification_report
```
