# 3RD STEP : HANDLE CATEGORICAL DATA

```python
#we convert categorical data into numeric bcoz machine only understands numbers
#create dataframe for object
df_cat=df.select_dtypes(object)
df_cat.head()
```

```python
#create dataframe for numeric
df_num=df.select_dtypes(['int64','float64'])
df_num.head()
```

```python
#convert categorical into numeric
from sklearn.preprocessing import LabelEncoder
```

```python
#apply labelencoder on df_cat
for col in df_cat:
    #create object
    le=LabelEncoder()
    df_cat[col]=le.fit_transform(df_cat[[col]]) #input always 2D
    
```

```python
df_cat.head()
```

```python
sns.heatmap(df_num.corr(),annot=True)
plt.show()
```

```python
#apply skewness on continuous numeric 
for col in df_num:
    s=df_num[col].skew()
    print(col,s)
```

```python
#check  neagative number
for col in df_num:
    print(col,' : ',df_num[col].unique())
```
